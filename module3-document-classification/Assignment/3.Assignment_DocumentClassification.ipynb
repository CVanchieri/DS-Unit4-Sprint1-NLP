```
# Dataset
from sklearn.datasets import fetch_20newsgroups

categories = ['alt.atheism',
              'talk.religion.misc']

data = fetch_20newsgroups(subset='train', categories=categories)
```

```
# Create Pipeline Components
vect = TfidfVectorizer(stop_words='english')
rfc = RandomForestClassifier()
```

```
# Define the Pipeline, pass a list of tuples
pipe = Pipeline([
                 #Vectorizer
                 ('vect', vect),
                 # Classifier
                 ('clf', rfc)
                ])

# The pipeline puts together a bunch fit_transform, fit_transform, then fit_predict.
```

```
# Grid search for both vectorizer and classifier
parameters = {
    'vect__max_df': ( 0.75, 1.0),
    'vect__min_df': (.02, .05),
    'vect__max_features': (500,1000),
    'clf__n_estimators':(5, 10,),
    'clf__max_depth':(15,20)
}

grid_search = GridSearchCV(pipe,parameters, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(data.data, data.target)
```

# TfidfVectorizer with GridSearchCV.

```
# Import Statements
from sklearn.pipeline import Pipeline
from sklearn.datasets import fetch_20newsgroups
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
# Create Pipeline Components

# set the vector with TfidfVectorizer.
vect = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
# set the random forest classifier.
rfc = RandomForestClassifier()
```

```
# create the pipeline.
pipe = Pipeline([
                 # vectorizer.
                 ('vect', vect),
                 # classifier.
                 ('clf', rfc)
                ])

# The pipeline puts together a bunch fit then transform, fit then predict. 
```

```
parameters = {
    'vect__max_df': ( 0.75, 1.0),
    'vect__min_df': (.02, .05)
    #'vect__max_features': (500,1000),
    #'clf__n_estimators':(5, 10,),
    #'clf__max_depth':(15,20)
}

grid_search = GridSearchCV(pipe,parameters, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(train['description'], train['category'])
```

```
grid_search.best_score_
```

```
# Predictions on test sample
pred = grid_search.predict(test['description'])
```

```
submission = pd.DataFrame({'id': test['id'], 'category':pred})
submission['category'] = submission['category'].astype('int64')
```

```
# Make Sure the Category is an Integer
submission.head()
```

# TruncatedSVD with GridSearchCV.

```
# import libraries.
from sklearn.decomposition import TruncatedSVD

# singular value decomposition, decomposes to reduce complexity.
svd = TruncatedSVD(n_components=100, # Just here for demo. 
                   algorithm='randomized',
                   n_iter=10)
```

```
params = { 
    'lsi__svd__n_components': [10,100,250],
    'lsi__vect__max_df':[.9, .95, 1.0],
    'clf__n_estimators':[5,10,20]
}
```

```
# LSI
lsi = Pipeline([('vect', vect), ('svd', svd)])


# create pipeline.
pipe = Pipeline([('lsi', lsi), ('clf', rfc)])
```

```
print(pipe)
```

```
# Fit
grid_search = GridSearchCV(pipe,params, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(train['description'], train['category'])
```

```
grid_search.best_score_
```

# TfidfVectorizer & RandomForest.

```
vect = TfidfVectorizer(stop_words='english')
clf = RandomForestClassifier()

lsi = Pipeline([('vect', vect), ('svd', svd)])
pipe = Pipeline([('lsi', lsi), ('clf', clf)])
```

```
# These parameters take a long time to run vs. defaults above
parameters = {
    'lsi__svd__n_components': [10,100,250],
    'lsi__vect__max_df':[.9, .95, 1.0],
    'clf__n_estimators':[5,10,20]
    
    # 'lsi__svd__n_components': [10,100,125],
    # 'lsi__vect__max_df': [.9, .95, 1.0],
    # 'lsi__vect__min_df': (.02, .05),
    # 'lsi__vect__max_features': (500,1000),
    # 'clf__n_estimators':[5, 10, 15, 20],
    # 'clf__max_depth':(5,10,15,20)
}

grid_search = GridSearchCV(pipe,parameters, cv=5, n_jobs=4, verbose=1)
grid_search.fit(train.description, train.category)
```

```
grid_search.best_score_
```

```
# Predictions on test sample
pred = grid_search.predict(test['description'])
```

```

submission = pd.DataFrame({'id': test['id'], 'category':pred})
submission['category'] = submission['category'].astype('int64')
```
